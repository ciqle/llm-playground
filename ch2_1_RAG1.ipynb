## Splitting Your Text into Chunks

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = TextLoader("./resources/novel_sample.txt") # or any other loader
docs = loader.load()

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
)
splitted_docs = splitter.split_documents(docs)
````

````python
splitted_docs
````

### split code languages

````python
from langchain_text_splitters import (
    Language,
    RecursiveCharacterTextSplitter,
)

PYTHON_CODE = """
def hello_world():
    print("Hello, World!")

# Call the function
hello_world()
"""
python_splitter = RecursiveCharacterTextSplitter.from_language( # from_language方法为特定语言创建一个实例
    language=Language.PYTHON, chunk_size=50, chunk_overlap=0
)
python_docs = python_splitter.create_documents([PYTHON_CODE])
````

````python
python_docs
````

````python
markdown_text = """
# LangChain

⚡ Building applications with LLMs through composability ⚡

## Quick Install

```bash
pip install langchain
```

As an open source project in a rapidly developing field, we are extremely open 
    to contributions.
"""

md_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.MARKDOWN, chunk_size=60, chunk_overlap=0
)
md_docs = md_splitter.create_documents(
    texts=[markdown_text], metadatas=[{"source": "https://www.langchain.com"}]
)
````

````python
md_docs
````

The metadata we passed in the second argument is attached to each resulting document, which allows you to track, for instance, where the document came from and where you can go to see the original.

我们传入的第二个参数中的元数据会附加到每个生成的文档中，这使得你可以追踪文档来源，并知道去哪里查看原始文件。

## Generating Text Embeddings

````python
# from langchain_openai import OpenAIEmbeddings
# model = OpenAIEmbeddings()

# in local use Ollama instead
from langchain_ollama import OllamaEmbeddings

model = OllamaEmbeddings(base_url="http://localhost:11434", model="nomic-embed-text")

embeddings = model.embed_documents(
    [
        "Hi there!",
        "Oh, hello!",
        "What's your name?",
        "My friends call me World",
        "Hello World!",
    ]
)

embeddings
````
